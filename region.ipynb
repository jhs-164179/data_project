# 제주도 시,읍,면 별 위도 경도 데이터 프레임 생성
- 최종 dts 인덱스

```python
import pandas as pd
import numpy as np
import json
from tqdm import tqdm

jeju_geo = json.load(open('./hangjeongdong_제주특별자치도.geojson', encoding='utf8'))
```

```python
key=['경도_Lng','위도_Lat','행정구역']
lng_lst=[]
lat_lst=[]
region_lst=[]

for geo in jeju_geo['features']:
    
    if geo['properties']['adm_nm'].find('동')==-1:
        region=geo['properties']['adm_nm']
    else:
        if geo['properties']['adm_nm'].find('제주시')==-1:
            region='제주특별자치도 서귀포시'
        else:
            region='제주특별자치도 제주시'
    
    for i in range(len(geo['geometry']['coordinates'])):
        for lng,lat in geo['geometry']['coordinates'][i][0]:
            region_lst.append(region)
            lng_lst.append(lng)
            lat_lst.append(lat)
```

```python
jeju_df=pd.DataFrame(dict(zip(key,[lng_lst,lat_lst,region_lst])))
jeju_df
```

```python
jeju_df.describe(include='all')
```

```python
jeju_df['경도_Lng']=jeju_df['경도_Lng'].map(lambda x:x-x%.001)
jeju_df['위도_Lat']=jeju_df['위도_Lat'].map(lambda x:x-x%.001)

jeju_df.head()
```

```python
jeju_df.describe(include='all')
```

```python
jeju_df=jeju_df.drop_duplicates()
jeju_df.describe(include='all')
```

```python
regions=jeju_df['행정구역'].unique()
regions
```

```python
# 서귀포시 = region
min1=jeju_df[np.logical_and(jeju_df['행정구역']=='제주특별자치도 서귀포시', jeju_df['위도_Lat']==np.sort(jeju_df[jeju_df['행정구역']=='제주특별자치도 서귀포시']['위도_Lat'].unique())[0])]['경도_Lng'].min()
max1=jeju_df[np.logical_and(jeju_df['행정구역']=='제주특별자치도 서귀포시', jeju_df['위도_Lat']==np.sort(jeju_df[jeju_df['행정구역']=='제주특별자치도 서귀포시']['위도_Lat'].unique())[0])]['경도_Lng'].max()
for num in np.arange(min1,max1,.001):
    num='{:.3f}'.format(num)
    print(num,type(num))
```

```python
key=['경도_Lng-위도_Lat','행정구역']
lng_lat_lst=[]
region_lst=[]

for region in tqdm(regions):
    lats=np.sort(jeju_df[jeju_df['행정구역']==region]['위도_Lat'].unique())
    for lat in lats:
        min_lng=jeju_df[np.logical_and(jeju_df['행정구역']==region, jeju_df['위도_Lat']==lat)]['경도_Lng'].min()
        max_lng=jeju_df[np.logical_and(jeju_df['행정구역']==region, jeju_df['위도_Lat']==lat)]['경도_Lng'].max()
        lngs=np.arange(min_lng,max_lng+.001,.001)
        for lng in lngs:
            lng='{:.3f}'.format(lng)
            lng_lat_lst.append(lng+'_'+str(lat))
            region_lst.append(region)
```

```python
df_final=pd.DataFrame(dict(zip(key,[lng_lat_lst,region_lst])))
df_final.head(10)
```
